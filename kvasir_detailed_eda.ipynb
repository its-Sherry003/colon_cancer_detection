# Detailed Exploratory Data Analysis (EDA) – Kvasir Dataset

This notebook performs a **detailed EDA** for the Kvasir gastrointestinal image dataset.

Analysis performed:
- Dataset structure inspection
- Class distribution
- Dataset imbalance check
- Sample image visualization
- Image resolution analysis
- Aspect ratio analysis
- Pixel intensity distribution
- RGB channel statistics
- Brightness distribution
- Duplicate image detection


```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from hashlib import md5

print('Libraries loaded')
```

## Dataset Path
Update this path to your dataset location.


```python
DATASET_PATH = 'kvasir-dataset'

classes = sorted(os.listdir(DATASET_PATH))
print('Classes:')
for c in classes:
    print(c)
```

## Count Images per Class


```python
class_counts = {}

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    class_counts[c] = len(os.listdir(folder))

df_counts = pd.DataFrame(list(class_counts.items()), columns=['Class','Count'])
df_counts
```

## Class Distribution Visualization


```python
plt.figure()
plt.bar(df_counts['Class'], df_counts['Count'])
plt.xticks(rotation=45)
plt.title('Class Distribution')
plt.ylabel('Number of Images')
plt.show()
```

## Dataset Imbalance Check


```python
total_images = df_counts['Count'].sum()

df_counts['Percentage'] = (df_counts['Count'] / total_images) * 100
df_counts
```

## Display Sample Images


```python
plt.figure()
i = 1

for c in classes[:6]:
    folder = os.path.join(DATASET_PATH, c)
    img_name = os.listdir(folder)[0]
    img_path = os.path.join(folder, img_name)

    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2,3,i)
    plt.imshow(img)
    plt.title(c)
    plt.axis('off')
    i += 1

plt.show()
```

## Image Resolution Analysis


```python
widths = []
heights = []

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    for img_name in os.listdir(folder)[:100]:
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path)
        h, w, _ = img.shape
        widths.append(w)
        heights.append(h)

print('Average Width:', np.mean(widths))
print('Average Height:', np.mean(heights))
print('Min Width:', np.min(widths))
print('Max Width:', np.max(widths))
print('Min Height:', np.min(heights))
print('Max Height:', np.max(heights))
```

## Width Distribution


```python
plt.figure()
plt.hist(widths)
plt.title('Image Width Distribution')
plt.xlabel('Width')
plt.ylabel('Frequency')
plt.show()
```

## Height Distribution


```python
plt.figure()
plt.hist(heights)
plt.title('Image Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

## Aspect Ratio Analysis


```python
aspect_ratios = [w/h for w,h in zip(widths,heights)]

plt.figure()
plt.hist(aspect_ratios)
plt.title('Aspect Ratio Distribution')
plt.xlabel('Width / Height')
plt.ylabel('Frequency')
plt.show()
```

## Pixel Intensity Distribution


```python
sample_pixels = []

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    for img_name in os.listdir(folder)[:20]:
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path,0)
        sample_pixels.extend(img.flatten())

plt.figure()
plt.hist(sample_pixels)
plt.title('Pixel Intensity Distribution')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

## RGB Channel Statistics


```python
r_vals, g_vals, b_vals = [], [], []

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    for img_name in os.listdir(folder)[:20]:
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path)

        b,g,r = cv2.split(img)

        r_vals.extend(r.flatten())
        g_vals.extend(g.flatten())
        b_vals.extend(b.flatten())

print('R mean:', np.mean(r_vals))
print('G mean:', np.mean(g_vals))
print('B mean:', np.mean(b_vals))
```

## Brightness Distribution


```python
brightness = []

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    for img_name in os.listdir(folder)[:20]:
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        brightness.append(np.mean(gray))

plt.figure()
plt.hist(brightness)
plt.title('Brightness Distribution')
plt.xlabel('Brightness')
plt.ylabel('Frequency')
plt.show()
```

## Duplicate Image Detection


```python
hashes = {}
duplicates = []

for c in classes:
    folder = os.path.join(DATASET_PATH, c)
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)

        with open(img_path,'rb') as f:
            file_hash = md5(f.read()).hexdigest()

        if file_hash in hashes:
            duplicates.append(img_path)
        else:
            hashes[file_hash] = img_path

print('Duplicate images found:', len(duplicates))
```
